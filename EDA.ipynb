#Multivariate EDA

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
titanic = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
```

```python
sns.scatterplot(x='total_bill', y='tip', hue = df['sex'], data = df)
```

```python
sns.scatterplot(x='total_bill', y='tip',hue = df['sex'], style = df['smoker'], data = df)
```

```python
sns.scatterplot(x='total_bill', y = 'tip', hue=df['sex'], style = df['smoker'], size = df['size'], data = df)
```

```python
sns.barplot(x='Pclass', y='Age', data=titanic)
```

```python
sns.barplot(x='Pclass', y='Age', hue=df['sex'], data=titanic)
```

```python
sns.boxplot(x='Pclass', y ='Age',data = titanic)
```

```python
sns.boxplot(x='Pclass', y='Age', hue=df['sex'], data=titanic)
```

```python
sns.barplot(x='Survived', y='Age', hue=df['sex'], data=titanic)
```

```python
sns.boxplot(x='Sex', y = 'Age', hue='Survived', data=titanic)
```

```python
sns.barplot(x='Pclass', y='Fare',data = titanic)
```
