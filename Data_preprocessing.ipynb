##De-duplicate
De-duplicate means removing all duplicate values. There is no need for duplicate values in data analysis. These values only affect the accuracy and efficency of the analysis result. To find duplicate values in the dataset, we will use a simple dataframe function, ie, duplicated().

```python
print(df.duplicated())
```

###Dataframe_name.drop_duplicates(subset = None, keep = 'first', inplace = False, ignore_index = False)
---

> subset (default = None)

Specifies which columns to consider for identifying duplicates
*   If None, all columns are used
*   If a list of columns is provided, duplicates are detected based on those columns

> keep (default = 'first')
Determines which duplicate row to keep
*   'first' : keeps the first occurrence and drops the rest
*   'last' : keeps the last occurrence and drops the other

False: Removes all duplicate rows

> Inplace (default = False)

*   If true, modifies the dataframe in place and returns None
*   If False, returns a new datafram with duplicates removed



> Ignore_index (default = False)


*   If true, resets the index in the resulting dataframe


```python
print(df.duplicated().sum())
```

##Dealing with outliers
(detection and removal of outliers)

```python
import sklearn
from sklearn.datasets import load_diabetes
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diabetes = load_diabetes()

column_name = diabetes.feature_names
df_diabetes = pd.DataFrame(diabetes.data, columns = column_name)

sns.boxplot(df_diabetes['bmi'])
plt.title("Boxplot of BMI")
plt.show()
```

```python
def removal_box_plot(df, column, threshold):
  removed_outliers = df[df[column] <= threshold]
  sns.boxplot(removed_outliers[column])
  plt.title(f"Box Plot without Outliers of {column}")
  plt.show()
  return removed_outliers
threshold_value = 0.12

no_outliers = removal_box_plot(df_diabetes, 'bmi', threshold_value)
```
