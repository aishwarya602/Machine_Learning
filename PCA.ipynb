##Practical - To perform dimentionality reduction using PCA


```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
#Generate some sample data
X = np.random.rand(100, 10)
```

```python
#Initialise PCA with the desired number of components
n_components = 2
pca = PCA(n_components=n_components)
```

```python
#fit PCA to the data and tranform it
X_pca = pca.fit_transform(X)
```

```python
print(X_pca)
```

```python
pca_one_component = PCA(n_components=1)
X_pca_one = pca_one_component.fit_transform(X)
```

```python
print("\nTransformed data with 1 component: ")
print(X_pca_one)
```

```python
#print the shape of the original and tranform data
print("\nShape of original data: ", X.shape)
print("Shape of transformed data: ", X_pca.shape)
```
