##Practical - 9 - Apply Naive-Byes classifier on dataset and analyze the prediction accuracy


```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
#split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify = y)
```

```python
#feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#create naive bayes model
model = GaussianNB()
model.fit(X_train_scaled, y_train)
```

```python
#make predictions
y_pred = model.predict(X_test_scaled)
```

```python
#model performance
accuracy = accuracy_score(y_test, y_pred)
print(f"\n Model Accuracy: {accuracy*100:.2f}%")
```

```python
#confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n", cm)
```

```python
from sklearn.metrics import classification_report

#classification
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
#confusion matrix using seaborn
plt.figure(figsize = (6,4))
sns.heatmap(cm, annot = True, fmt = 'd', cmap = 'Blues', xticklabels = iris.target_names, yticklabels = iris.target_names)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Naive Byes Confusion Matrix')
plt.show()
```

```python
#analyze prediction
comparision = pd.DataFrame({"Actual":[iris.target_names[i] for i in y_test], "Predicted":[iris.target_names[i] for i in y_pred]})
print("\nSample comparision of actual vs predicted labels:")
print(comparision.head(10))
```
